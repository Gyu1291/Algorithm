```python
import pandas as pd
import numpy as np
import math
```

```python
sheet_ad=pd.read_excel('E:/다운로드/SMS_dataset.xlsx')
```

```python
#Ad에 대해 작업한 것이다.
dt=[]
for i in range(len(sheet_ad)):
    dt.append(sheet_ad.values[i][0])
for i in range(len(dt)):
    dt[i]=dt[i].lower()
    dt[i]=dt[i].replace("."," ").replace(","," ").replace("-"," ").replace("/"," ").replace("http","link ").replace("?"," ").replace("!"," ").replace('"',' ').replace("'"," ").replace("^"," ").replace("입니다"," ").replace("에서"," ")
LK=[]
for i in range(len(dt)):
    LK.append(dt[i].split(' '))
LCK=[]
for i in range(len(LK)):
    for j in range(len(LK[i])):
        LCK.append(LK[i][j])

Lth3=[]
for i in range(len(LCK)):
    if len(LCK[i])>3:
        Lth3.append(LCK[i])
    else:
        pass
    
df=pd.DataFrame(Lth3,columns=['Text'])

words=df.groupby(['Text'])['Text'].count()
words=pd.DataFrame(words)
words=words.rename(columns={'Text':'count'})
words=words.reset_index()
words['add one']=words['count']+1
words['prob']=words['add one']/words['add one'].sum()
words['log(p)']=np.log(words['prob'])
words
```

```python
sheet_other=pd.read_excel('E:/다운로드/SMS_dataset.xlsx', sheet_name='Others')
```

```python
dt2=[]
for i in range(len(sheet_other)):
    dt2.append(sheet_other.values[i][0])
for i in range(len(dt2)):
    dt2[i]=dt2[i].lower()
    dt2[i]=dt2[i].replace("."," ").replace(","," ").replace("-"," ").replace("/"," ").replace("http","link ").replace("?"," ").replace("!"," ").replace('"',' ').replace("'"," ").replace("^"," ").replace("("," ").replace(")"," ").replace("입니다"," ").replace("에서"," ")
LK2=[]
for i in range(len(dt2)):
    LK2.append(dt2[i].split(' '))
    
LCK2=[]
for i in range(len(LK2)):
    for j in range(len(LK2[i])):
        LCK2.append(LK2[i][j])

Lth3_2=[]
for i in range(len(LCK2)):
    if len(LCK2[i])>3:
        Lth3_2.append(LCK2[i])
    else:
        pass
    
df2=pd.DataFrame(Lth3_2,columns=['Text'])

words2=df2.groupby(['Text'])['Text'].count()
words2=pd.DataFrame(words2)
words2=words2.rename(columns={'Text':'count'})
words2=words2.reset_index()
words2['add one']=words2['count']+1
words2['prob']=words2['add one']/words2['add one'].sum()
words2['log(p)']=np.log(words2['prob'])
words2
```

```python
sheet_test=pd.read_excel('E:/다운로드/SMS_dataset.xlsx', sheet_name='Testset')
```

```python
test=[]
for i in range(len(sheet_test)):
    test.append(sheet_test.values[i][2])
for i in range(len(test)):
    test[i]=test[i].lower()
    test[i]=test[i].replace("."," ").replace(","," ").replace("-"," ").replace("/"," ").replace("http","link ").replace("?"," ").replace("!"," ").replace('"',' ').replace("'"," ").replace("^"," ").replace("입니다"," ").replace("에서"," ")
test_split=[]
for i in range(len(test)):
    test_split.append(test[i].split(' '))

test_split
```

```python
app_prob=[[] for cols in range(len(test_split))]
for i in range(len(test_split)):
    for j in range(len(test_split[i])):
        if len(test_split[i][j])>3:
            if(words[words['Text'].isin([test_split[i][j]])]['log(p)'].empty):
                app_prob[i].append(math.log(1/words['add one'].sum()))
            else:
                app_prob[i].append(words[words['Text'].isin([test_split[i][j]])]['log(p)'].values[0])
        else:
            app_prob[i].append(0.0)
```

```python
S=[]
for i in range(len(app_prob)):
    S.append(sum(app_prob[i]))
S
```

```python
other_prob=[[] for cols in range(len(test_split))]
for i in range(len(test_split)):
    for j in range(len(test_split[i])):
        if len(test_split[i][j])>3:
            if(words2[words2['Text'].isin([test_split[i][j]])]['log(p)'].empty):
                other_prob[i].append(math.log(1/words2['add one'].sum()))
            else:
                other_prob[i].append(words2[words2['Text'].isin([test_split[i][j]])]['log(p)'].values[0])
        else:
            other_prob[i].append(0.0)
```

```python
T=[]
for i in range(len(other_prob)):
    T.append(sum(other_prob[i]))
T
```

```python
N=[]
for i in range(len(S)):
    if S[i]>T[i]:
        N.append('Ad')
    else:
        N.append('Other')
N
```

```python
Answer=[]
for i in range(len(S)):
    Answer.append(sheet_test.values[i][1])
Answer

def accuracy(a,b):
                      
    cnt = 0
    for i in range(len(a)):
        if a[i]==b[i]:
            cnt+=1
    c = cnt/len(a)
    print(f'{c*100}% Accuracy!')

accuracy(N,Answer)
```
