```python
import pandas as pd
import numpy
```

```python
df=pd.read_excel('E:/다운로드/cluster.xlsx')
```

```python
df
```

```python
pt=df.pivot_table(index = ['Offer #'], columns = ['Customer_Last_Name'], aggfunc=len, fill_value=0)
pt
```

```python
pt_T=pt.T
```

```python
pt_T
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmeans=KMeans(n_clusters=4).fit(pt_T)
label=kmeans.labels_
sil_coeff=silhouette_score(pt_T , label , metric='euclidean')
```

```python
sil_coeff
```

```python
kmeans=KMeans(n_clusters=4, max_iter=1000).fit(pt_T)
pt_T['clusters']=kmeans.labels_
```

```python
pt_T
```

```python
sil={}
for i in range(3,6):
    kmeans=KMeans(n_clusters=i).fit(pt_T)
    label=kmeans.labels_
    sil_coeff=silhouette_score(pt_T,label,metric='euclidean')
    sil[i]=sil_coeff
```

```python
sil
```

```python
wine_df=pd.read_excel('E:/다운로드/cluster.xlsx', sheet_name='OfferInformation')
```

```python
wine_df['clus_0']=pt_T[pt_T.clusters==0].sum(axis=0)[:-1]
wine_df['clus_1']=pt_T[pt_T.clusters==1].sum(axis=0)[:-1]
wine_df['clus_2']=pt_T[pt_T.clusters==2].sum(axis=0)[:-1]
wine_df['clus_3']=pt_T[pt_T.clusters==3].sum(axis=0)[:-1]
```
