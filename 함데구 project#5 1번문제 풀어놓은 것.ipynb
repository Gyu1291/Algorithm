```python
################################# Problem 1.1 #################################
class BinHeap:
    def __init__(self):
        self.heap_list = [0]
        self.current_size = 0
    
    ### CODE HERE ###
    """binary heap의 기능에 대한 설명이 없어서 강의안에 있는 minheap을 기준으로 하여 maxheap을 구현하였다."""
    def perc_up(self, i): 
        while i // 2 > 0: # root에 다다를 때까지
            if self.heap_list[i] > self.heap_list[i // 2]: #자식노드에 부모노드보다 큰 값이 들어가있으면
                tmp = self.heap_list[i // 2] 
                self.heap_list[i // 2] = self.heap_list[i] 
                self.heap_list[i] = tmp
            i = i // 2            
            #child_node가 parent보다 크면 둘을 바꿔주는 시행을 한다.
            
    def insert(self, k): #insert는 perc_up을 이용하므로 O(logn)의 complexity를 가진다.
        self.heap_list.append(k) 
        self.current_size = self.current_size + 1 
        self.perc_up(self.current_size)      
        
    def perc_down(self, i):
        while (i * 2) <= self.current_size: #가장 끝단 노드가 아닌 경우 while문이 동작
            mc = self.max_child(i)
            if self.heap_list[i] < self.heap_list[mc]: #child가 더 크면 parent와 child를 바꿔준다.
                tmp = self.heap_list[i]
                self.heap_list[i] = self.heap_list[mc]
                self.heap_list[mc] = tmp
            i = mc
            
    def max_child(self, i): #하나의 parent아래에 있는 두개의 child중 큰 값의 index를 반환한다.
        if i * 2 + 1 > self.current_size: #right child가 없는경우 left child반환
            return i * 2
        else:
            if self.heap_list[i * 2] > self.heap_list[i * 2 + 1]:
                return i * 2 
            else:
                return i * 2 + 1        
            
    def del_max(self): # max_heap이므로 root가 가장 클 것이고 root를 지워주는 시행이다.
        ret_val = self.heap_list[1] 
        self.heap_list[1] = self.heap_list[self.current_size] #마지막 노드에 있는 값을 root에 넣고
        self.current_size = self.current_size - 1 #size를 하나 줄여준다. 
        self.heap_list.pop() #마지막 노드에 있는 값을 없애고
        self.perc_down(1) #root에 있는 값을 다시 마지막 노드로 보낸다. 이 경우 처음 root에 있던 값이 삭제된다.
        return ret_val #최초에 root에 들어있던 값을 반환해준다.
    
    
    def build_heap(self, a_list): # input으로 들어온 a_list를 heap의 구조로 만드는 것이다.
        i = len(a_list) // 2
        self.current_size = len(a_list)
        self.heap_list = [0] + a_list[:]
        while (i > 0):
            self.perc_down(i) 
            i=i-1       
    #################


################################################################################
```

```python
################################# Problem 1.2 #################################

def find_median(lst):
    """ 
    This function finds and "prints" the medians of the all sublist, lst[0:i] using BinHeap.
    
    Args:
        lst (list) : The list of random integers.
    
    Returns:
        None
        
    """
    ### CODE HERE ###
    max_heap=BinHeap() #이 heap은 maxheap의 기능을 하며 median보다 작은 값들을 넣어 줄 것이다.
    min_heap=BinHeap() #이 heap는 minheap의 기능을 하며 median보다 큰 값들을 넣어 줄 것이다.
    """max_heap은 기준점보다 작은 값들이 들어가서 max_heap의 역할을 할 것이다. min_heap은 class는 max_heap이지만
    내가 insert할 때 -를 붙여서 원소들의 부호를 바꿔 줄 것이고 이러면 min_heap처럼 사용할 수 있다."""
    
    """우리가 구하는 중간값을 a라고 하면 a보다 작은 원소들은 max_heap에 a보다 큰 원소들은 min_heap에 들어가고 a는 max_heap의
    root가 되도록 설정하여, max_heap의 root를 반환하면 중간값을 반환할 수 있도록 할 것이다."""
    
    for i in range(len(lst)): #i가 1~n이고 lst[0:i]이면 결국 lst[i-1]이 매번 heap에 추가된다고 보면 되므로 i는 0~n-1로 설정
        if i==0: #맨 처음에는 그냥 max_heap에 원소를 insert하고 root에 있는 값을 그대로 출력해준다.
            max_heap.insert(lst[i]) 
            print(max_heap.heap_list[1]) #max_heap의 root에는 median이 있고 이를 출력해준다.
        else: #max_heap에 이미 원소가 있을 경우
            if lst[i]>max_heap.heap_list[1]: #새로 들어오는 값이 max_heap의 root보다 크면
                min_heap.insert(-lst[i]) #min_heap에 넣어주되 이 때 부호를 바꿔서 넣어준다.
            else:
                max_heap.insert(lst[i]) #새로 들어오는 값이 기존 median보다 크지 않을 경우 그냥 max_heap에 넣어준다.
                
            """이후 두개의 heap의 size를 맞춰주면 max_heap의 root에 mid값이 오게 된다. size는 max_heap과 min_heap이 동일하거나
            max_heap의 사이즈가 1 더 커야한다.(중앙값은 max_heap에 저장하도록 했으므로) 이 균형이 맞게 될 때까지 원소가 많은
            쪽에서 del_max()를 시행해 반대쪽으로 insert해주면 된다. 그런데 이 때 for문의 매 시행마다
            heap의 균형을 맞춰주므로 특정 시점에서 둘의 size는 커봐야 2차이가 나고,
            균형이 맞지 않는 상태에서도 원소를 한번만 옮겨주면 된다.
            즉 while문을 필요로 하지 않고 조건문만 써주면 된다.
            """
            
            """max_heap에는 median값이 들어가므로 전체 숫자가 홀수개일 때 minheap보다 size가 1 더 클수 있으며
            min_heap은 어떤 경우에도 max_heap보다 size가 크면 안된다. 이를 어길 경우 root에 있는 숫자를 상대편으로 옮긴다"""
            if max_heap.current_size>min_heap.current_size+1: #max_heap이 큰 경우
                k=max_heap.del_max() #max_heap에서 root를 빼고
                min_heap.insert(-k)#이를 부호를 바꿔서 minheap으로 넣는다.
            elif min_heap.current_size>max_heap.current_size: #min_heap이 큰 경우
                k=min_heap.del_max() #min_heap의 root에서 값을 빼는데, 이때 반환하는 값은 처음 insert된 값에 -를 붙인 값이다.
                max_heap.insert(-k) #따라서 이를 max_heap에 넣어줄 때 다시 -를 붙여서 원래의 값을 되찾도록 한다.
                
            print(max_heap.heap_list[1]) #max_heap의 root에는 median이 있고 이를 출력해준다.
            
    return
            
    #################

################################################################################
```

```python
class Node:
    def __init__(self, key, left_child=None, right_child=None, parent=None):
        self.key = key
        self.left_child = left_child
        self.right_child = right_child
        self.parent = parent
        
    def has_left_child(self): 
        return self.left_child
    
    def has_right_child(self): 
        return self.right_child
    
    def is_left_child(self):
        return self.parent and self.parent.left_child == self
    
    def is_right_child(self):
        return self.parent and self.parent.right_child == self
    
    def is_root(self):
        return not self.parent
    
    def is_leaf(self):
        return not (self.right_child or self.left_child)
    
    def has_any_children(self):
        return self.right_child or self.left_child
    
    def has_both_children(self):
        return self.right_child and self.left_child


class SplayTree:
    def __init__(self):
        self.root = None
    
    def splay(self, node):
        """ 
        This method splays the nodes.
    
        Args:
            node (Node) : The node to splay.
    
        Returns:
            None
        
        """
        ### CODE HERE ###
        """splay 특징. x가 root가 될 때가지 하므로 while문 안에서 돌아야 함
        rotate를 잘 짜면 zig-zig랑 zig-zag는 rotate과정을 두 번 수행하는 코드로 짜면 됨."""
        
        """함수안의 함수로 rotate left와 rotate right를 구현하는게 더 좋아보이는데 일단은 그냥짜두고 시간남으면 해보자
        굳이 정의가 안들어간다면 실행시간은 이 코드가 더 빠를 것 같음"""
        while node.parent!=None: #node가 root가 될 때까지 반복해야하므로, 또한 node가 이미 root이면 실행되지 않음
            if node.parent.parent==None: # 조부모가 없을 때=> parent가 root일 때=> rotate해야함
                if node.parent.left_child==node: #rotate_left
                    node.parent.left_child=node.right_child
                    node.right_child=node.parent
                    node.parent.parent=node
                    node.parent=None #어차피 grandparent가 None이기 때문
                    
                else: #rotate_right
                    node.parent.right_child=node.left_child
                    node.left_child=node.parent
                    node.parent.parent=node
                    node.parent=None #어차피 grandparent가 None이기 때문
                    
            else: #parent가 root가 아니므로 최소한 grandparent가 있는 경우이다.
                if node.is_left_child() and node.parent.is_left_child(): #left_zig_zig
                    node.parent.parent.left_child=node.parent.right_child
                    node.parent.right_child=node.parent.parent
                    temp=node.parent.parent #기존 grandparent를 temp로 둠
                    node.parent.parent=node.parent.parent.parent #node.parent의 parent를 node.parent.parent.parent로 바꿈
                    temp.parent=node.parent #기존 grandparent였던 노드의 parent를 node.parent로 바꿈
                    
                    #여기까지가 rotate(P)고 이 아래는 left_child인 X에 대해 rotate(X)
                    node.parent.left_child=node.right_child
                    node.right_child=node.parent
                    temp2=node.parent
                    node.parent=node.parent.parent
                    temp2.parent=node
                    
                    
                elif node.is_right_child() and node.parent.is_right_child():#right_zig-zig
                    node.parent.parent.right_child=node.parent.left_child
                    node.parent.left_child=node.parent.parent
                    temp=node.parent.parent #기존 grandparent를 temp로 둠
                    node.parent.parent=node.parent.parent.parent #node.parent의 parent를 node.parent.parent.parent로 바꿈
                    temp.parent=node.parent #기존 grandparent였던 노드의 parent를 node.parent로 바꿈
                    
                    #여기까지가 rotate(P)고 이 아래는 right_child인 X에 대해 rotate(X)
                    node.parent.right_child=node.left_child
                    node.left_child=node.parent
                    temp2=node.parent
                    node.parent=node.parent.parent
                    temp2.parent=node
                    
                    
                elif node.is_right_child() and node.parent.is_left_child(): #left_zigzag
                    node.parent.right_child=node.left_child
                    node.left_child=node.parent
                    temp=node.parent
                    node.parent=node.parent.parent
                    temp.parent=node
                    
                    node.parent.left_child=node.right_child
                    node.right_child=node.parent
                    temp2=node.parent
                    node.parent=node.parent.parent
                    temp2.parent=node
                    
                    
                elif node.is_left_child() and node.parent.is_right_child(): #right_zigzag
                    node.parent.left_child=node.right_child
                    node.right_child=node.parent
                    temp=node.parent
                    node.parent=node.parent.parent
                    temp.parent=node
                    
                    node.parent.right_child=node.left_child
                    node.left_child=node.parent
                    temp2=node.parent
                    node.parent=node.parent.parent
                    temp2.parent=node
        
        self.root=node        
            
        #################
    
    def insert(self, key): #마지막 노드의 자식으로만 달아줘도 BST만족 가능
        """ 
        This method inserts the key value to splay tree.
    
        Args:
            key : The key value to insert
    
        Returns:
            None
        
        """
        ### CODE HERE ###
        current=self.root
        previous=self.root
        new_node=Node(key,None,None,None)
        if self.root==None: #아무것도 없는 상태
            self.root=new_node #그냥 new_node를 root로 설정해주면 된다.
        else:
            while current: #while반복문이 들어가지만 모든 노드를 탐색하지 않고 크기를 비교하며 선택해 내려가므로 O(logn)
                if previous.key==key: #current!=None인데 key값이 동일한 노드를 발견하면 right_child에 insert
                    temp=previous.right_child
                    previous.right_child=new_node
                    new_node.parent=previous
                    new_node.right_child=temp
                    temp.parent=new_node
                    return #이 경우 return 으로 함수실행 종료
                    
                elif current.key>key: #current.key가 key보다 크면 currnet를 왼쪽으로 내려감
                    previous=current
                    current=current.left_child
                else: #current.key가 key보다 크지 않은 경우 current는 right로 내려감. 같은 경우라면 다음 반복에서 위의 if에 걸림
                    previous=current
                    current=current.right_child
                    
            if previous.key>key: #멈추지 않고 while문을 탈출한 경우 previous는 leaf node임.
                previous.left_chlid=new_node
                new_node.parent=previous
            else:
                previous.right_child=new_node
                new_node.parent=previous
                
            self.splay(new_node) #끝단에 추가시킨 이후 splay해주면 된다.
            #################

    def find(self, key):
        """ 
        This method finds the node have the specified key.
    
        Args:
            key : The key value to find.
    
        Returns:
            Bool
        
        """
        ### CODE HERE ###
        current=self.root
        previous=self.root
        if not current: #self.root가 None이면 그냥 시작부터 종료함.
            return False
        
        while current:
            if previous.key==key:
                self.splay(previous)
                return True
            elif current.key>key:
                prvious=current
                current=current.left_child
            else:
                previous=current
                current=current.right_child
                
        if previous.key==key:
            self.splay(previous)
            return True
        else:
            self.splay(previous)
            return False
        #################

    def delete(self, key):
        """ 
        This method delete the node has the specified key.
    
        Args:
            key : The key value to delete.
    
        Returns:
            None
        
        """
        ### CODE HERE ###
        a=self.find(key) #못찾은 경우 알아서 splay하고 끝난다. 찾아서 True를 반환하면 아래 코드를 실행한다. 
        if a:
            current=self.root
            if not current.has_any_child:
                self.root=None
            elif not current.has_both_child:
                if current.has_left_child:
                    self.root=current.left_child
                    current.left_child.parent=None
                else:
                    self.root=current.right_child
                    current.right_child.parent=None
            else: #child가 모두 있는 경우
                temp=current.right_child
                max_node=current.left_child
                while max_node.right_child!=None:
                    max_node=max_node.right_child #왼쪽 subtree에서 key값이 가장 큰 노드는 항상 맨오른쪽에 위치한다.
                self.splay(max_node)
                max_node.right_child=temp
                temp.parent=max_node
                
        #################
    
    def pre_order(self, node=None):
        """ 
        Do a preorder traversal.
    
        Args:
            node : The target node.
    
        Returns:
            None
        
        """
        ### CODE HERE ###
        if node!=None:
            print(node.key)
            pre_order(node.left_child)
            pre_order(node.right_child)
        #################
    
    def post_order(self, node=None):
        """ 
        Do a postorder traversal.
    
        Args:
            node : The target node.
    
        Returns:
            None
        
        """
        ### CODE HERE ###
        if node!=None:
            post_order(node.left_child)
            post_order(node.right_child)
            print(node.key)
        #################
    
    def in_order(self, node=None):
        """ 
        Do an inorder traversal.
    
        Args:
            node : The target node.
    
        Returns:
            None
        
        """
        ### CODE HERE ###
        if node!=None:
            inorder(node.left_child)
            print(node.key)
            inorder(node.right_child)
        #################
```

```python
def print_tree(tree_node, level=0):
    if isinstance(tree_node, SplayTree):
        tree_node = tree_node.root
        
    if tree_node == None:
        return
    
    if level != 0:
        print("           ", end='')
        if tree_node.parent.left_child == tree_node:
            print("             "*(level - 1) + '_____(L)' + '[{:2d}]'.format(tree_node.key))
        else:
            print("             "*(level - 1) + '_____(R)' + '[{:2d}]'.format(tree_node.key))
    else:
        print('\n----- Printing tree structure -----')
        print('(Root)[{:2d}]'.format(tree_node.key))
        
    print_tree(tree_node.right_child, level + 1)
    print_tree(tree_node.left_child, level + 1)
```
