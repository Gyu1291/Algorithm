```python
N, K = map(int, input().split())
coin=[int(input()) for _ in range(N)]
cnt=0

for i in range(1,N+1):
    if coin[N-i]>K:
        continue
    else:
        while K >= coin[N-i]:
            K = K - coin[N-i]
            cnt+=1
        #나는 while문을 돌렸는데 여기서 나눗셈을 시행해서 몫과 나머지로 접근하는게
        #효율적이다
            
print(cnt)
```

```python
#주유소 문제-쉽게 풀어서 기분좋다
n=int(input())
distance=list(map(int,input().split()))
price=list(map(int,input().split()))

total_price=0

for i in range(n-1):
    if i==0:
        total_price += distance[0] * price[0]
        lower_cost = price[0]
    else:
        lower_cost=min(lower_cost,price[i])
        total_price+=distance[i]*lower_cost
        
print(total_price)
```

```python
A, B = map(int, input().split())
number=1
for i in range(B):
    number *= A-i
    number /= i+1
    
print(number)
#import math 한다음에 factorial 쓰면 된다
```

```python
from math import factorial
C, D = map(int, input().split())
print(factorial(C)/(factorial(D)*factorial(C-D)))
```

```python
n, k = map(int, input().split())
dp = [[0] * 1 for i in range(1001)]
dp[1].append(1)
for i in range(2, 1001):
    for j in range(1, i + 1):
        if j == 1:
            dp[i].append(1)
        elif j == i:
            dp[i].append(1)
        else:
            dp[i].append(dp[i - 1][j - 1] + dp[i - 1][j])
print(dp[n + 1][k + 1] % 10007)
```
